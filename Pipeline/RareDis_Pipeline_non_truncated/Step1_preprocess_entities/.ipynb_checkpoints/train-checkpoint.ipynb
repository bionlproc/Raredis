```python
import json

train = []
for line in open('RareDis_Pipeline/Datasets_with_only_gold_entities/train.json', 'r'):
    train.append(json.loads(line))

for i in range(len(train)):
    train[i]['doc_key'] = train[i].pop('doc')
    train[i]['sentences'] = train[i].pop('tokens')
    train[i]['ner'] = train[i].pop('entities')

    train[i]['ner_modified'] = train[i]['ner'].copy()
    train[i]['sentences_modified'] = train[i]['sentences'].copy()

    train[i].pop('start')
    train[i].pop('end')
    # train[i].pop('text')

# Convert number strings to number lists
def str2intlist(s):
    l = []
    l.append(int(s.split(',')[0]))
    l.append(int(s.split(',')[1]))
    return l

for i in range(len(train)):

    l = []

    for j in range(len(train[i]['ner'])):
        l_temp = []
        for k in range(len(train[i]['ner'][j]['span'])):
            l_temp.append(str2intlist(train[i]['ner'][j]['span'][k]))
        l_temp.append(train[i]['ner'][j]['type'])
        l.append(l_temp)

    train[i]['ner'] = l

for i in range(len(train)):

    l = []

    for j in range(len(train[i]['ner_modified'])):

        l_temp = []

        for k in range(len(train[i]['ner_modified'][j]['span'])):
            l_temp.append(str2intlist(train[i]['ner_modified'][j]['span'][k]))

        l_temp.append(train[i]['ner_modified'][j]['type'])

        l.append(l_temp)

    train[i]['ner_modified'] = l
```

```python
# Remove discontinuous entities with more than 2 fragments

for i in range(len(train)):
    
    indices_to_remove = set()
    
    for j in range(len(train[i]['ner_modified'])):
        if len(train[i]['ner_modified'][j]) >= 4:
            indices_to_remove.add(j)
            
    new_list = [sublist for idx, sublist in enumerate(train[i]['ner_modified']) if idx not in indices_to_remove]
    
    train[i]['ner_modified'] = new_list
```

```python
num_non_overlapped = 0
num_overlapped = 0

for doc in range(len(train)):
    
    for j in range(len(train[doc]['ner_modified'])):  
        if len(train[doc]['ner_modified'][j]) == 3:

            if 0 < j < len(train[doc]['ner_modified'])-1:

                t = 0
                for k in range(len(train[doc]['ner_modified'][j])-1):
                    if train[doc]['ner_modified'][j-1][-2][1] < train[doc]['ner_modified'][j][k][0] < train[doc]['ner_modified'][j+1][0][0] and train[doc]['ner_modified'][j-1][-2][1] < train[doc]['ner_modified'][j][k][1] < train[doc]['ner_modified'][j+1][0][0]:
                        t += 1

                if t == len(train[doc]['ner_modified'][j]) - 1:
                    print(f"The {j}th distinuous entity in doc_{doc} is non-overlapped.")
                    num_non_overlapped += 1
                elif train[doc]['ner_modified'][j][0][0] > train[doc]['ner_modified'][j-1][-2][1]:
                    print(f"The {j}th distinuous entity in doc_{doc} is overlapped and overlap with the LATTER entity.")
                    num_overlapped += 1
                else:
                    print(f"The {j}th distinuous entity in doc_{doc} is overlapped and overlap with the PREVIOUS entity.")
                    num_overlapped += 1

            if j == 0:

                t = 0
                for k in range(len(train[doc]['ner_modified'][j])-1):
                    if train[doc]['ner_modified'][j][k][0] < train[doc]['ner_modified'][j+1][0][0] and train[doc]['ner_modified'][j][k][1] < train[doc]['ner_modified'][j+1][0][0]:
                        t += 1

                if t == len(train[doc]['ner_modified'][j]) - 1:
                    print(f"The {j}th distinuous entity in doc_{doc} is non-overlapped.")
                    num_non_overlapped += 1
                else:
                    print(f"The {j}th distinuous entity in doc_{doc} is overlapped and overlap with the LATTER entity.")
                    num_overlapped += 1

            if j == len(train[doc]['ner_modified'])-1:       

                t = 0
                for k in range(len(train[doc]['ner_modified'][j])-1):
                    if train[doc]['ner_modified'][j-1][-2][1] < train[doc]['ner_modified'][j][k][0] and train[doc]['ner_modified'][j-1][-2][1] < train[doc]['ner_modified'][j][k][1]:
                        t += 1

                if t == len(train[doc]['ner_modified'][j]) - 1:
                    print(f"The {j}th distinuous entity in doc_{doc} is non-overlapped.")
                    num_non_overlapped += 1
                else:
                    print(f"The {j}th distinuous entity in doc_{doc} is overlapped and overlap with the PREVIOUS entity.")
                    num_overlapped += 1
                    
            print('\n')
```

```python
# Modify sentences with 2-fragment discontinuous entities

for doc in range(len(train)):

    x = train[doc]['sentences_modified']
    y = train[doc]['ner_modified']

    for j in range(len(train[doc]['ner_modified'])):
        if len(train[doc]['ner_modified'][j]) == 3:  # 2-fragment discontinuous entities

            if 0 < j < len(train[doc]['ner_modified']) - 1:  # Neither the first entity nor the last entity

                t = 0
                for k in range(len(train[doc]['ner_modified'][j]) - 1):
                    if train[doc]['ner_modified'][j - 1][-2][1] < train[doc]['ner_modified'][j][k][0] < \
                            train[doc]['ner_modified'][j + 1][0][0] and train[doc]['ner_modified'][j - 1][-2][1] < \
                            train[doc]['ner_modified'][j][k][1] < train[doc]['ner_modified'][j + 1][0][0]:
                        t += 1

                if t == len(train[doc]['ner_modified'][j]) - 1:  # Confirm the entity is non-overlapped

                    x = (x[:train[doc]['ner_modified'][j][0][0]]
                         + x[train[doc]['ner_modified'][j][0][0]:train[doc]['ner_modified'][j][0][1] + 1]
                         + x[train[doc]['ner_modified'][j][1][0]:train[doc]['ner_modified'][j][1][1] + 1]
                         + x[train[doc]['ner_modified'][j][1][1] + 1:])  # Modify sentences

                    train[doc]['sentences_modified'] = x

                    gap = y[j][1][0] - y[j][0][1] - 1  # Gap is between the two fragments of the discontinuous entity

                    y[j] = [[y[j][0][0], y[j][1][1] - gap], y[j][-1]]  # Modify offsets for the current discontinuous entity

                    for l in range(j + 1, len(y)):  # Modify offsets for all other entities *after* the discontinuous entity
                        for k in range(len(y[l]) - 1):
                            y[l][k][0] -= gap
                            y[l][k][1] -= gap

                    train[doc]['ner_modified'] = y

                    
                else:
                    print(f"The {j}th distinuous entity in doc_{doc} is overlapped.")
                    

            if j == 0:

                t = 0
                for k in range(len(train[doc]['ner_modified'][j]) - 1):
                    if train[doc]['ner_modified'][j][k][0] < train[doc]['ner_modified'][j + 1][0][0] and \
                            train[doc]['ner_modified'][j][k][1] < train[doc]['ner_modified'][j + 1][0][0]:
                        t += 1

                if t == len(train[doc]['ner_modified'][j]) - 1:

                    x = (x[:train[doc]['ner_modified'][j][0][0]]
                         + x[train[doc]['ner_modified'][j][0][0]:train[doc]['ner_modified'][j][0][1] + 1]
                         + x[train[doc]['ner_modified'][j][1][0]:train[doc]['ner_modified'][j][1][1] + 1]
                         + x[train[doc]['ner_modified'][j][1][1] + 1:])  # Modify sentences

                    train[doc]['sentences_modified'] = x
                    
                    

                    gap = y[j][1][0] - y[j][0][1] - 1  # Gap is between the two fragments of the discontinuous entity

                    y[j] = [[y[j][0][0], y[j][1][1] - gap], y[j][-1]]  # Modify offsets for the current discontinuous entity

                    for l in range(j + 1, len(y)):  # Modify offsets for all other entities *after* the discontinuous entity
                        for k in range(len(y[l]) - 1):
                            y[l][k][0] -= gap
                            y[l][k][1] -= gap

                    train[doc]['ner_modified'] = y

                else:
                    print(f"The {j}th distinuous entity in doc_{doc} is overlapped.")
                    

            if j == len(train[doc]['ner_modified']) - 1:  # The discontinuous entity is the last entity

                t = 0
                for k in range(len(train[doc]['ner_modified'][j]) - 1):
                    if train[doc]['ner_modified'][j - 1][-2][1] < train[doc]['ner_modified'][j][k][0] and \
                            train[doc]['ner_modified'][j - 1][-2][1] < train[doc]['ner_modified'][j][k][1]:
                        t += 1

                if t == len(train[doc]['ner_modified'][j]) - 1:
                    x = (x[:train[doc]['ner_modified'][j][0][0]]
                         + x[train[doc]['ner_modified'][j][0][0]:train[doc]['ner_modified'][j][0][1] + 1]
                         + x[train[doc]['ner_modified'][j][1][0]:train[doc]['ner_modified'][j][1][1] + 1]
                         + x[train[doc]['ner_modified'][j][1][1] + 1:])  # Modify sentences

                    train[doc]['sentences_modified'] = x

                    gap = y[j][1][0] - y[j][0][1] - 1  # Gap is between the two fragments of the discontinuous entity

                    y[j] = [[y[j][0][0], y[j][1][1] - gap], y[j][-1]]  # Modify offsets for the current discontinuous entity

                    train[doc]['ner_modified'] = y
                else:
                    print(f"The {j}th distinuous entity in doc_{doc} is overlapped.")
                
            print('\n')
```

```python
def check_overlap(interval1, interval2):
    # Extract the start and end points of each interval
    start1, end1 = interval1
    start2, end2 = interval2

    # Check for overlap
    if end1 < start2 or end2 < start1:
        return False  # Overlapping intervals
    else:
        return True  # Non-overlapping intervals
```

```python
for doc in range(len(train)):

    x = train[doc]['sentences_modified']
    y = train[doc]['ner_modified']

    for j in range(len(train[doc]['ner_modified'])):

        if len(train[doc]['ner_modified'][j]) == 3:  # 2-fragment discontinuous entities
            
            interval_1st = train[doc]['ner_modified'][j][0]
            interval_2nd = train[doc]['ner_modified'][j][1]

            switch = 0

            for idx in range(len(train[doc]['ner_modified'])):

                if idx == j:
                    pass

                else:

                    if len(train[doc]['ner_modified'][idx]) == 2: # Continuous entities
                        interval_cont = train[doc]['ner_modified'][idx][0] 
                        
                        if check_overlap(interval_2nd, interval_cont):
                            switch = 1
                            break
                        
                        elif check_overlap(interval_1st, interval_cont):
                            switch = 3
                            break
                               
                        
                    elif len(train[doc]['ner_modified'][idx]) == 3: # 2-fragment discontinuous entities
                        interval_discont_1st = train[doc]['ner_modified'][idx][0]
                        interval_discont_2nd = train[doc]['ner_modified'][idx][1]
                        
                        
                        if check_overlap(interval_2nd, interval_discont_2nd) is True and check_overlap(interval_1st, interval_discont_1st) is False:
                            switch = 2
                            break


                        elif check_overlap(interval_1st, interval_discont_1st) is True and check_overlap(interval_2nd, interval_discont_2nd) is False:
                            switch = 4
                            break               
                        

            if switch == 1 or switch == 2:

                # print('Find a overlapped entity with switch 1:', train[doc]['ner_modified'][j])

                x = (x[:train[doc]['ner_modified'][j][0][1]+1]  
                + x[train[doc]['ner_modified'][j][1][0]:train[doc]['ner_modified'][j][1][1]+1] 
                + x[train[doc]['ner_modified'][j][0][1]+1:])

                len_2nd_frag = train[doc]['ner_modified'][j][1][1] - train[doc]['ner_modified'][j][1][0] + 1

                cut = train[doc]['ner_modified'][j][0][1]

                y[j] = [[y[j][0][0], y[j][0][1] + len_2nd_frag], y[j][-1]]

                for l in range(len(y)): 
                    
                    
                    if l == j:
                        pass

                    else:
                    

                        for k in range(len(y[l]) - 1):
                            if y[l][k][0] > cut and y[l][k][1] > cut:
                                y[l][k][0] += len_2nd_frag
                                y[l][k][1] += len_2nd_frag
                            else:
                                pass

                train[doc]['sentences_modified'] = x

                train[doc]['ner_modified'] = y 

            elif switch == 0 or switch == 3 or switch == 4:

                # print('Find a overlapped entity with switch 0:', train[doc]['ner_modified'][j])
                
                len_1st_frag = y[j][0][1] - y[j][0][0] + 1

                x = (x[:train[doc]['ner_modified'][j][1][0]] 
                     + x[train[doc]['ner_modified'][j][0][0]:train[doc]['ner_modified'][j][0][1]+1] 
                     + x[train[doc]['ner_modified'][j][1][0]:])

                cut = train[doc]['ner_modified'][j][1][0]

                y[j] = [[y[j][1][0],y[j][1][1]+len_1st_frag], y[j][-1]]

                for l in range(len(y)):         
                    
                    if l == j:
                        pass

                    else:
                    

                        for k in range(len(y[l]) - 1):

                            if y[l][k][0] >= cut and y[l][k][1] >= cut:
                                y[l][k][0] += len_1st_frag
                                y[l][k][1] += len_1st_frag
                            else:
                                pass

                train[doc]['sentences_modified'] = x

                train[doc]['ner_modified'] = y 
                

                
            else:
                print(f'Switch is {switch}')
```

# Preprocess relations for any doc

```python
train_re = []
for line in open('train_re.json', 'r'):
    train_re.append(json.loads(line))
```

```python
for doc in range(len(train)):
    train[doc]['relations'] = []
    
    L = train[doc]['ner']
    
    L_removed = []
    for idx in range(len(train[doc]['ner'])):
        if len(train[doc]['ner'][idx]) > 3:
            L_removed.append(idx)
            
    # Create a dictionary to store the names and associated lists
    dict_L = {f"T{i+1}": L[i] for i in range(len(L))}
    
    # Remove the elements in L_removed from the dictionary
    for idx in L_removed:
        del dict_L[f"T{idx+1}"]
        
    # Convert the dictionary items to a list
    items = list(dict_L.items())
    
    new_dict_L = {}
    for k in range(len(train[doc]['ner_modified'])):
        new_dict_L[items[k][0]] = train[doc]['ner_modified'][k]
        
        
    if 'relations' in train_re[doc]:
        
        for i in range(len(train_re[doc]['relations'])):

            for key, value in train_re[doc]['relations'][i].items():

                if value[0] in new_dict_L and value[1] in new_dict_L:
                    rel_temp = []
                    rel_temp += (new_dict_L[value[0]][0] + new_dict_L[value[1]][0])
                    rel_temp.append(key)
                else:
                    pass   

            train[doc]['relations'].append(rel_temp)
            
    else:
        print(f'Doc {doc} does not contain any relation.')
```
